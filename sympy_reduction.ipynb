```python
import sympy as sp

# Symbols and unknown functions
x, alpha, ell, w0 = sp.symbols(
    "x alpha ell w0",
    nonzero=True
)

f = sp.Function("f")
g = sp.Function("g")
V = sp.Function("V")

L = ell * (ell + 1)

# Equations are stored as expressions equal to zero

# Radial equation
RE = (
    sp.diff(f(x), x)
    + w0 * (alpha * g(x) - x * sp.diff(g(x), x))
)

# Timelike equation
TE = (
    3 * alpha * f(x)
    - 3 * x * sp.diff(f(x), x)
    + w0 * sp.diff(g(x), x)
    + w0 / x * (2 * g(x) - L * V(x))
)

# Polar equation
PE = (
    alpha * w0 * V(x)
    - w0 * x * sp.diff(V(x), x)
    + f(x) / x
)

# Solve RE and PE for g' and V'
rules = sp.solve(
    [RE, PE],
    [
        sp.diff(g(x), x),
        sp.diff(V(x), x)
    ],
    dict=True
)[0]

# Substitute those derivative rules into TE ; produces the algebraic constraint involving g and V.
C = sp.together(
    TE.subs(rules)
).as_numer_denom()[0]

# Differentiate the constraint and replace g' and V'
Cp = sp.together(
    sp.diff(C, x).subs(rules)
).as_numer_denom()[0]

C = sp.expand(C)
Cp = sp.expand(Cp)

## find the multiple of C that cancels g and V
multiplier = sp.simplify(
    Cp.coeff(g(x)) / C.coeff(g(x))
)

ode_lhs = sp.expand(
    -(Cp - multiplier * C)
)

# collect by f'', f', and f
ode_lhs = sp.collect(
    ode_lhs,
    [
        sp.diff(f(x), x, 2),
        sp.diff(f(x), x),
        f(x)
    ]
)

ode = sp.Eq(ode_lhs, 0)

sp.pprint(ode)
```

```python
assert not ode_lhs.has(g(x))
assert not ode_lhs.has(V(x))

print("Multiplier used:", multiplier)
```
